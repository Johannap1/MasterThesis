## GP training

```python
train_x_1 = train_x[[2]].reshape(-1,1)
train_y_1 = train_y[[2]].reshape(-1,1)

train_x_2 = train_x[[2,3]]
train_y_2 = train_y[[2,3]]

train_x_3 = train_x[[0,2,3]]
train_y_3 = train_y[[0,2,3]]

train_x_4 = train_x[[0,2,3,4]]
train_y_4 = train_y[[0,2,3,4]]

train_x_5 = train_x[[0,1,2,3,4]]
train_y_5 = train_y[[0,1,2,3,4]]
```

```python
X = train_x_2
y = train_y_2
```

```python
# We will use the simplest form of GP model, exact inference
class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood):
        super(ExactGPModel, self).__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ZeroMean()
        self.covar_module = gpytorch.kernels.ScaleKernel(gpytorch.kernels.RBFKernel())

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)

# initialize likelihood and model
noise_level = 0.0
noise = noise_level*torch.ones(X.shape[0])
likelihood = gpytorch.likelihoods.FixedNoiseGaussianLikelihood(noise=noise, learn_additional_noise=False)
model = ExactGPModel(X, y, likelihood)

```

```python
# this is for running the notebook in our testing framework
training_iter = 20


# Find optimal model hyperparameters
model.train()
likelihood.train()

# Use the adam optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)  # Includes GaussianLikelihood parameters

# "Loss" for GPs - the marginal log likelihood
mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)

for i in range(training_iter):
    # Zero gradients from previous iteration
    optimizer.zero_grad()
    # Output from model
    output = model(X)
    # Calc loss and backprop gradients
    loss = -mll(output, y)
    loss.backward()
    optimizer.step()
```

```python
# Get into evaluation (predictive posterior) mode
model.eval()
likelihood.eval()

# Test points are regularly spaced along [0,1]
# Make predictions by feeding model through likelihood
with torch.no_grad(), gpytorch.settings.fast_pred_var():
    test_x = torch.linspace(0, 1, 51)
    observed_pred = likelihood(model(true_x))
```

```python
with torch.no_grad():
    # Initialize plot
    f, ax = plt.subplots(figsize=(12, 6))

    # Get upper and lower confidence bounds
    lower, upper = observed_pred.confidence_region()
    # Plot training data as black stars
    ax.scatter(train_x, train_y, s=200, marker='o',linestyle="None", facecolors='none', edgecolors=(12/255, 35/255, 64/255), alpha = 0.5, label='_nolegend_')
    ax.plot(X, y, color=(12/255, 35/255, 64/255), marker='X',linestyle="None", markersize=10)
    # Plot predictive means as blue line
    ax.plot(true_x.numpy(), observed_pred.mean.numpy(), color=(0/255, 166/255, 214/255) )
    # Shade between the lower and upper confidence bounds
    ax.fill_between(true_x.numpy(), lower.numpy(), upper.numpy(), color=(0/255, 166/255, 214/255), alpha=0.5)
    ax.plot(true_x, np.zeros(true_x.shape),'gray', alpha = 0.5,label='_nolegend_')
    ax.set_ylim([-3, 3])
    ax.legend(['Observed data','Mean', 'Confidence'])
    ax.set_xlim([0,10])
    plt.axis('off')
    ax.set_facecolor("gray")
    ax.add_artist(ax.patch)
    ax.patch.set_zorder(-1)
    ax.patch.set_alpha(0.1)
    name = '/home/johanna/thesis_plots/gp_2'
    if not name==None:
        plt.savefig(name,  bbox_inches="tight")
```

```python
def plot_path(X=None,  res=8, title=None, name=None):
    fig, ax = plt.subplots(1, 1)
    plt.axis('equal')
    plt.grid(linestyle=':')
    plt.plot(X[::res, 0], X[::res, 1], 'k--')
    plt.title(title, fontsize=16)
    plt.xlabel('x [m]', fontsize=18)
    plt.ylabel('y [m]', fontsize=18)
    plt.rc('xtick', labelsize=18) 
    plt.rc('ytick', labelsize=18) 
    plt.ioff()
    plt.autoscale()
    if not name==None:
        plt.savefig(name,  bbox_inches="tight")
    plt.show()
```
